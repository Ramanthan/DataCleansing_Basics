```python
# creating function to do Null check on the dataset
def Null_check(df):
    d=pd.DataFrame(df.isnull().sum()/len(df)).reset_index()
    d=d.rename(columns={"index": "Columns",0:"Null_score"})
    d['Null_score']=round(d['Null_score']*100,2)
    c=d.loc[d['Null_score'] != 0,:].reset_index()
    l=(c['Columns'].to_list())
    global x
    x=c.reset_index()
    if len(c)>0:
        print("Data set contains NULL values for {0} and below are the columns and its NULL value proportion".format(len(l)))
        print("\n")
        print(c[['Columns','Null_score']])
        print("\n")
        print("We need to handle the NULL values by using statistical method or by Domain expertise")
    else:
        print("Nice,No null value present in Dataset")
        
        
        
```

# Removing Data - Column Level 

When the columns in the dataset contains NULL value more than 25% of its total values, we need to remove the columns directly from the data set

```python
def Remove_NULL_25(df):
    Null_cols=[i for i in df.columns if round(df[i].isnull().sum()/len(df)*100,2) > 25]
    print("Below columns contains more than 25 % of NULL values and can drop it from the dataset\n")
    print(Null_cols)
    df.drop(Null_cols,axis=1,inplace=True)
    print("Now we have to impute values based on the value distribution")
```

```python
Remove_NULL_25(df)
```
